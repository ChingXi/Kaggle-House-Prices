```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_train = pd.read_csv('train.csv')
df_train
```

```python
df_train.describe()
```

```python
x = df_train['YearBuilt']
y = df_train['SalePrice']
data = pd.concat([x, y], axis=1)
data.plot.scatter(x = 'YearBuilt', y = 'SalePrice');
```

```python
x = df_train['GrLivArea']
y = df_train['SalePrice']
data = pd.concat([x, y], axis=1)
data.plot.scatter(x = 'GrLivArea', y = 'SalePrice');
```

```python
sns.boxplot(x=df_train['OverallQual'], y=df_train['SalePrice']);
```

```python
sns.boxplot(x=df_train['OverallCond'], y=df_train['SalePrice']);
```

```python
corrmat = df_train.corr()
corrmat
```

```python
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corrmat, cmap='coolwarm', vmax = 0.8)
```

```python
corr = np.corrcoef((df_train['SalePrice'], df_train['OverallQual'], df_train['TotalBsmtSF'], df_train['GrLivArea'], df_train['GarageCars']))
corr
```

```python
cols = corrmat.nlargest(10, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```
